## Drawing with OpenCV Python Computer Vision

### 1. Drawing Lines with Python

```python
import numpy as np
img1 = np.zeros((512,512,3), np.uint8)
cv.line(img1,(0,0),(511,511),(255,0,0),5)
cv.imshow("line", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#cv.line(img1,(0,0),(511,511),(255,0,0),5)
```

### 2. Drawing rectangles in Python

```python
cv.rectangle(img1, (384,0),(510,128),(0,255,0),3)
cv.imshow("Rectangle", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 3. Drawing Circles in Python

```python
cv.circle(img1, (447,63), 63,(0,0,255),-1)
cv.imshow("Circle", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 4. Drawing ellipses in Python

```python
cv.ellipse(img1, (256,256), (100,50), 0, 0, 180, 100, -1)
cv.imshow("ellipses", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 5. Drawing Polygon in Python

```python
pts = np.array([[10,5],[50,70],[250,360],[400,400]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img1, [pts], True, (0,255,255))
cv.imshow("Polygon", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Putting Text in Images

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img1, "OpenCV", (10,500), font, 4,(255,255,23),2,cv.LINE_AA)
cv.imshow("Text_Image", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

# BGR color scheme and Region of Interest

```python
import cv2 as cv
img2 = cv.imread("scene.jpeg")
```

```python
y,x = 100, 50
# Reading color values at y,x positions
(b,g,r) = img2[y,x]
# Printing color values
b,g,r
```

```python
# Setting pixel color to red; BGR scheme
img2[y,x] = (0,0,255)
```

```python
# region of Interest at (x,y) of dimensions 50 x 50 
region_of_interest = img2[y:y+50, x:x+50]
cv.imshow("image", img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imshow("ROI", region_of_interest)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Setting pixels to new color
region_of_interest[:,:] = (66,22,78)
cv.imshow("New_Image", img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Detecting Edges in Python Computer Vision

```python
import numpy as np
img4 = cv.imread('scene.jpeg')
cv.imwrite("edges.jpg", cv.Canny(img4, 602, 315))
```

```python
cv.imshow("Edges", cv.imread('edges.jpg'))
cv.waitKey(0)
cv.destroyAllWindows()
```
